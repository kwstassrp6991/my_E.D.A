```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Customer Database Analysis
- Exploratory data analysis on a dataset of 100 customers.
- Libraries used: pandas, matplotlib

```python
df = pd.read_csv("customers-100.csv")
```

```python
df.info()
```

- Check if we have Nan values in the file 

```python
df.isnull().sum()
```

- The top 1 Country With the most subscribers 

```python
top =df["Country"].value_counts().head(1)
```

- The top 10 countries with the most subscribers 

```python
top_ten = df["Country"].value_counts().head(10)
```

- Remake the column of the date in dtype datetime
- How many subscribers per year 

```python
df["Subscription Date"] = pd.to_datetime(df["Subscription Date"])
df["Year"] = df["Subscription Date"].dt.year
df["Month"] = df["Subscription Date"].dt.month
df["Day"] = df["Subscription Date"].dt.day
new_customers_of_the_year = df["Year"].value_counts()
```

- Most common Names

```python
most_common_names= df["First Name"].value_counts().head(10)
```

```python
## Analyze every subscriber's Website 
## security 
cyber_code = df["Website"]
cyber_code1 = tuple(cyber_code.values)

http_count = 0
https_count = 0
for i in cyber_code1:
    if "https" in i:
        https_count += 1
    elif "http" in i:
        http_count += 1
```

- Create the charts (subplots)

```python
## put the whole Charts in  subplots
figure,axes = plt.subplots(2,2,figsize=(15,8.5))

#axes 0,0
axes[0,0].barh(top_ten.index,top_ten.values, align='center')
axes[0,0].set_title("Top 10 Countries")
axes[0,0].tick_params(axis ="y",color="lightgray",size=0,labelsize=6.5)

#axes 0,1
axes[0,1].plot(new_customers_of_the_year.sort_values().sort_index(),marker= "o")
axes[0,1].set_title("Subscriptions Per Year")
axes[0,1].set_ylabel("Subscribers")
axes[0,1].set_xticks(new_customers_of_the_year.index)

#axes 1,0
axes[1,0].bar(most_common_names.index, most_common_names.values)
axes[1,0].set_title("Most Common Names")
axes[1,0].set_ylabel("Times appear")
axes[1,0].tick_params(axis="x",labelsize=6.5)

#axes 1,1
codes = ["http","https"]
axes[1,1].pie([http_count, https_count],
                                        labels=codes,
                                        autopct = "%1.1f%%",
                                        explode = (0,0.1))
axes[1,1].legend(codes,bbox_to_anchor=(0.9, 1))
```

```python
plt.show()
```
